```python
import pandas as pd
docs = pd.read_json('docs_with_embeddings_non_outliers.json', lines=True)
reviews  =pd.read_json('filtered_dataset/selected_users.json', lines=True)

# Contruindo data frame com usuario | produto consumido | nota
consumes = reviews.join(docs.set_index("asin"), on="asin", how="inner")
consumes = consumes[['reviewerID', 'asin', 'vec', 'topic_probs_vec', 'overall', 'topic']]
consumes
```

```python
reviews_split = reviews.join(docs.set_index("asin"), on="asin", how="inner")
user_review_count = reviews_split["reviewerID"].value_counts()

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.hist(user_review_count, bins=range(0, user_review_count.max() + 10, 10), edgecolor='k', alpha=0.7)
plt.title('Distribuição de número de reviews por usuário')
plt.xlabel('Número de reviews')
plt.ylabel('Quantidade de usuários')
plt.yscale('log')  # opcional: útil se a distribuição for muito desbalanceada
plt.show()
```

```python
reviews_split['review_count'] = reviews_split['reviewerID'].map(user_review_count)
reviews_split

reviews_split = reviews_split[(reviews_split['review_count'] <= 180) & (reviews_split['review_count'] >= 20)]
reviews_split['reviewerID'].value_counts()
```

```python
import numpy as np

train = consumes

def mark_train_test(group, frac=0.70):
    n = len(group)
    n_train = int(np.floor(frac * n))
    idx_train = group.sample(n=n_train, random_state=42).index
    group['set'] = np.where(group.index.isin(idx_train), 'train', 'test')
    return group

def mean_vec(series):
    arrs = [np.asarray(v) for v in series]
    return np.mean(np.stack(arrs, axis=0), axis=0)

train = train.groupby('reviewerID', group_keys=False).apply(mark_train_test)

consumes = train

test = train[train['set'] == 'test']
train = train[train['set'] == 'train']

consumes
```

```python
# Train possui os consumos de treino e Test possui os consumos de teste

# Para cada usuario, sortear um igual numero de negativos

# reviewerID | asin | topic_probs_vec 

reviewers = reviews['reviewerID'].unique()
reviewers

to_concat = []

for uid in reviewers:
    r_positives = train[train['reviewerID'] == uid]['asin']
    r_negatives = docs[~docs['asin'].isin(consumes[consumes['reviewerID'] == uid]['asin'])][['asin', 'topic_probs_vec']]
    r_negatives = r_negatives.sample(r_positives.shape[0]*5)
    r_negatives['reviewerID'] = uid
    to_concat.append(r_negatives)

negatives = pd.concat(to_concat)
negatives
```

```python
from sklearn.linear_model import LogisticRegression

def penalize_user_embedding(w_u: np.ndarray,
                            pos_item_topic_vectors: np.ndarray,
                            alpha: float = 0.5) -> np.ndarray:
    
    # desvio-padrão por tópico nos positivos do usuário
    sigma_pos = pos_item_topic_vectors.std(axis=0, ddof=0)
    w_adj = w_u - alpha * sigma_pos
    return w_adj
    
X_pos_df = train
X_pos_df['example'] = 1

X_pos_df = X_pos_df[['reviewerID', 'topic_probs_vec', 'example']]

X_neg_df = negatives
X_neg_df['example'] = 0
X_neg_df = X_neg_df[['reviewerID', 'topic_probs_vec', 'example']]

X_df = pd.concat([X_pos_df, X_neg_df])

clf = LogisticRegression(penalty='l2', max_iter=1000)

# ReviewerID | embedding
reviewers_embeddings = []

for uid in reviewers:
    X_uid_df = X_df[X_df['reviewerID'] == uid]
    y = X_uid_df['example'].values
    X = np.vstack(X_uid_df['topic_probs_vec'].values)
    
    clf.fit(X, y)
    
    w_u = clf.coef_.ravel()
    b_u = clf.intercept_[0]
    w_u = w_u / (np.linalg.norm(w_u))
    
    embedding = w_u
    penalized_embedding = penalize_user_embedding(
        w_u,
        np.vstack(X_uid_df[X_uid_df['example'] == 1]['topic_probs_vec'].values),
        alpha=0.5,
    )

    reviewers_embeddings.append([uid, embedding, penalized_embedding])

user_embeddings = pd.DataFrame(reviewers_embeddings, columns=['ReviewerID', 'embedding', 'penalized_embedding'])
```

```python
def cos_sim(a, b):
    if np.linalg.norm(a) == 0 or np.linalg.norm(b) == 0:
        return 0.0
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))


scores_df = pd.merge(test_df, user_embeddings, left_on='reviewerID', right_on='ReviewerID', how='inner')
scores_df['topic_probs_vec_norm'] = scores_df['topic_probs_vec'].apply(lambda v: v / np.linalg.norm(v))
scores_df['score'] = scores_df.apply(lambda row: cos_sim(row['embedding'], row['topic_probs_vec_norm']), axis=1)
scores_df['penalized_score'] = scores_df.apply(lambda row: cos_sim(row['penalized_embedding'], row['topic_probs_vec_norm']), axis=1)
scores_df = scores_df[['reviewerID', 'embedding', 'penalized_embedding', 'topic_probs_vec_norm', 'score', 'penalized_score', 'example']]
scores_df
```

```python
import numpy as np
import pandas as pd

def ndcg_at_k(df, score_field, k=10):
    users = df['reviewerID'].unique()
    ndcgs = []

    for uid in users:
        user_df = df[df['reviewerID'] == uid].copy()

        if user_df[user_df['example'] == 1].shape[0] < k:
            continue

        # define relevância binária
        user_df['relevance'] = user_df['example']

        # ordena pelas pontuações previstas
        user_df = user_df.sort_values(by=score_field, ascending=False)

        # pega o top-k do ranking
        top_k = user_df.head(k)
        rels = top_k['relevance'].to_numpy()

        # calcula DCG@k
        dcg = np.sum((2**rels - 1) / np.log2(np.arange(2, len(rels) + 2)))

        # calcula IDCG@k (ranking ideal)
        ideal_rels = np.sort(user_df['relevance'].to_numpy())[::-1][:k]
        idcg = np.sum((2**ideal_rels - 1) / np.log2(np.arange(2, len(ideal_rels) + 2)))

        ndcg = dcg / idcg if idcg > 0 else 0
        ndcgs.append(ndcg)

    return np.mean(ndcgs) if ndcgs else 0

mean_ndcg = ndcg_at_k(scores_df, 'score', k=1)
p_mean_ndcg = ndcg_at_k(scores_df, 'penalized_score', k=5)
print(f"nDCG@100 médio: {mean_ndcg:.4f}")
print(f"nDCG@100 médio: {p_mean_ndcg:.4f}")
```
