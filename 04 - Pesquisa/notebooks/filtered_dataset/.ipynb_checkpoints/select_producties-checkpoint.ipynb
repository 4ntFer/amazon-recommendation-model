```python
import pandas as pd

df = pd.read_json("../raw_dataset/Movies_and_TV_5.json",
                  lines=True,
                  nrows=2_000_000)

df = df.drop(columns=["style", "image"])
```

```python
df
```

```python
prod = pd.read_json("../filtered_dataset/filtered_meta.json", lines=True)
df = df[df["asin"].isin(prod['asin'])]

df
```

```python
user_review_count = df["reviewerID"].value_counts()

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.hist(user_review_count, bins=range(0, user_review_count.max() + 10, 10), edgecolor='k', alpha=0.7)
plt.title('Distribuição de número de reviews por usuário')
plt.xlabel('Número de reviews')
plt.ylabel('Quantidade de usuários')
plt.yscale('log')  # opcional: útil se a distribuição for muito desbalanceada
plt.show()
```

```python
user_review_count.mean()
```

```python
df['review_count'] = df['reviewerID'].map(user_review_count)
df
```

```python
split = df[(df['review_count'] >= 50) & (df['review_count'] <= 400)]
split["reviewerID"].value_counts()
```

```python
split["reviewerID"].value_counts().mean()
```

```python
df = split
```

```python
user_review_count = df["reviewerID"].value_counts()

user_review_count
```

```python
df.to_json("selected_users.json", orient="records", lines=True, force_ascii=False)
```
