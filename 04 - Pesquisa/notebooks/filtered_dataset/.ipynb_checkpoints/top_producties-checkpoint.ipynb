```python
import pandas as pd

df = pd.read_json("../raw_dataset/Movies_and_TV_5.json",
                  lines=True,
                  nrows=1_000_000)

df = df.drop(columns=["style", "image"])
```

```python
df
```

```python
prod = pd.read_json("../filtered_dataset/filtered_meta.json", lines=True)
df = df[df["asin"].isin(prod['asin'])]

df
```

```python
user_review_count = df["reviewerID"].value_counts()

user_review_count
```

```python
user_review_count.median()
```

```python
contagens = df["reviewerID"].value_counts()

valores_frequentes = contagens[contagens > 10].index

df = df[df["reviewerID"].isin(valores_frequentes)]
```

```python
user_review_count = df["reviewerID"].value_counts()

user_review_count
```

```python
user_review_count.median()
```

```python
prod_counts = df['asin'].value_counts()
prod_counts
```

```python
prod_counts.mean()
```

```python
valores_frequentes = prod_counts.head(1000).index
valores_frequentes
```

```python
df = df[df['asin'].isin(valores_frequentes)]
df
```

```python
df.to_json("filtered_reviews.json", orient="records", lines=True, force_ascii=False)
```

```python
selected_producties = prod[prod['asin'].isin(valores_frequentes)]
selected_producties
```

```python
selected_producties.to_json("selected_producties.json", orient="records", lines=True, force_ascii=False)
```
