```python
import pandas as pd

meta_json = "../raw_dataset/meta_Movies_and_TV.json"

df = pd.read_json(meta_json, lines=True)
```

```python
df.sort_values(by="description", ascending=True)
```

```python
print(df.dtypes)
```

```python
df.isnull().sum()
```

```python
def contar_invalidos(df):
    invalid_counts = {}
    for col in df.columns:
        def invalido(x):
            if isinstance(x, str) and x.strip() == "":
                return True
            return False

        invalid_counts[col] = df[col].apply(invalido).sum()

    return pd.Series(invalid_counts, name="Valores Inválidos")


resumo = contar_invalidos(df)

resumo_df = pd.DataFrame({
    "Total": len(df),
    "Inválidos": resumo,
    "% Inválidos": (resumo / len(df) * 100).round(2)
})

print(resumo_df)
```

```python
contagem_vazios = df["description"].apply(lambda x: isinstance(x, (list, tuple)) and len(x) > 0).sum()

print("Descrições não vazias:", contagem_vazios)
```

```python
df = df[df["description"].apply(lambda x: len(x) > 0)]
```

```python
print(f"Linhas no dataframe antes do corte: {len(df)}")
df = df.explode("description", ignore_index=True)
df = df[df["description"]!=""]
df = df.dropna(subset=["description"])
df = df.groupby("asin", as_index=False).agg({"title": "first", "category": "first", "description": list, "feature": "first", "rank": "title", "rank": "first", "brand": "first", "details": "first"})
print(f"Linhas no dataframe após do corte: {len(df)}")
```

```python
contagem_vazios = df["category"].apply(lambda x: isinstance(x, (list, tuple)) and len(x) == 0).sum()

print("Categorias Vazias vazias:", contagem_vazios)
```

```python
contagem_vazios = df["feature"].apply(lambda x: isinstance(x, (list, tuple)) and len(x) == 0).sum()

print("Categorias Vazias vazias:", contagem_vazios)
```

```python
# Usar Asin como id

# Limpar titulos vazios
df = df[df["title"] != ""]
```

```python
df = df.explode("category", ignore_index=True)
df["category"].value_counts()
```

```python
# Limpar categorias genericas (Kindle Store, Kindle eBooks)
df = df.explode("category", ignore_index=True)
```

```python
print(f"Linhas no dataframe antes do corte: {len(df)}")
# Limpar categorias genericas (Kindle Store, Kindle eBooks)
df = df.explode("category", ignore_index=True)
df = df[~df["category"].isin(['Movies & TV', 'Movies', 'Genre for Featured Categories'])]
print(f"Linhas no dataframe após do corte: {len(df)}")

df = df.groupby("asin", as_index=False).agg({"title": "first", "category": list, "description": 'first', "feature": 'first', "rank": "title", "rank": "first", "brand": "first"})
```

```python
# Tratando coluna de rank

df["rank"] = (
    df["rank"]
    .str.extract(r"([\d,]+)")[0]        
    .str.replace(",", "", regex=False)  
    .astype("Int64")                    
)
df
```

```python
df.to_json("filtered_meta.json", orient="records", lines=True, force_ascii=False)
```
