```python
import pandas as pd
docs = pd.read_json('docs_with_embeddings_non_outliers.json', lines=True)
reviews =pd.read_json('filtered_dataset/selected_users.json', lines=True)
```

```python
reviews_split = reviews.join(docs.set_index("asin"), on="asin", how="inner")
user_review_count = reviews_split["reviewerID"].value_counts()

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.hist(user_review_count, bins=range(0, user_review_count.max() + 10, 10), edgecolor='k', alpha=0.7)
plt.title('Distribuição de número de reviews por usuário')
plt.xlabel('Número de reviews')
plt.ylabel('Quantidade de usuários')
plt.yscale('log')  # opcional: útil se a distribuição for muito desbalanceada
plt.show()
```

```python
reviews_split['review_count'] = reviews_split['reviewerID'].map(user_review_count)
reviews_split

reviews_split = reviews_split[(reviews_split['review_count'] <= 180) & (reviews_split['review_count'] >= 20)]
reviews_split['reviewerID'].value_counts()
```

```python
# Contruindo data frame com usuario | produto consumido | nota
consumes = reviews_split
consumes = consumes[['reviewerID', 'asin', 'vec', 'topic_probs_vec', 'overall', 'topic']]
consumes
```

```python
users = pd.read_json('reviewer_lr_embeddings.json', lines=True)
users = users.rename(columns={'ReviewerID': 'reviewerID','embedding': 'topic_probs_vec'})
users
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import numpy as np
import umap

Y = np.vstack(users['topic_probs_vec'].apply(np.asarray).to_numpy())

pca = PCA(n_components=2, random_state=0)
X2 = pca.fit_transform(Y)

plt.figure(figsize=(7,6))
plt.scatter(X2[:,0], X2[:,1], s=5, alpha=0.6)
plt.title("Distribuição (PCA 2D)")
plt.xlabel("PC1"); plt.ylabel("PC2")
plt.show()
```

```python
pca = PCA(n_components=2, random_state=0)
X2 = pca.fit_transform(Y)

plt.figure(figsize=(7,6))
plt.scatter(X2[:,0], X2[:,1], s=5, alpha=0.6)
plt.title("Distribuição topics probabilities (PCA 2D)")
plt.xlabel("PC1"); plt.ylabel("PC2")
plt.show()
```

```python
reducer = umap.UMAP(
    n_neighbors=15, min_dist=0.1,
    metric="cosine", random_state=0
)
X2 = reducer.fit_transform(Y)

plt.figure(figsize=(7,6))
plt.scatter(X2[:,0], X2[:,1], s=5, alpha=0.6)
plt.title("Distribuição topics probabilities (UMAP 2D)")
plt.xlabel("Dim1"); plt.ylabel("Dim2")
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=20, n_init="auto", random_state=0)

y_labels = kmeans.fit_predict(Y)

users['user_topics_cluster'] = y_labels

cluster_sizes = (
    users["user_topics_cluster"]
    .value_counts()
    .rename_axis("user_topics_cluster")
    .reset_index(name="cluster_size")
)

users = users.merge(cluster_sizes, on="user_topics_cluster", how="left")

users
```

```python
consumes = consumes.join(users[['reviewerID', 'user_topics_cluster', 'cluster_size']].set_index("reviewerID"), on="reviewerID", how="left")
consumes
```

```python
categories = docs.explode('category')
categories = categories[['asin', 'category']]
categories = categories.join(consumes.set_index("asin"), on="asin", how="left")
categories_count = categories.groupby(['user_topics_cluster', 'category']).size().reset_index(name='count')
categories = categories.merge(categories_count, on=['user_topics_cluster', 'category'], how="left")
categories = categories.rename(columns={'count': 'cluster_category_consume'})

cluster_total = (
    categories.groupby("user_topics_cluster")["reviewerID"]
    .size()
    .reset_index(name="cluster_total_consume")
)
categories = categories.merge(cluster_total, on="user_topics_cluster", how="left")

categories['cluster_category_consume_proportion'] = 100*categories['cluster_category_consume']/categories['cluster_total_consume']

categories
```

```python
distributions = categories[['category', 'user_topics_cluster', 'cluster_category_consume_proportion']].groupby(['category', 'user_topics_cluster'], as_index=False).agg({"cluster_category_consume_proportion": "first"})
distributions
```

```python
categories["cluster_category_consume_proportion"].plot(kind="kde", figsize=(8,5), title="Densidade de Consumo por Categoria")
# tudo que estiver abaixo de 15% pode ser considerado um consumo baixo,
# tudo que estiver acima disso posso considerar um consumo regular ou alto
```

```python
high_consuming = categories[['asin', 'category', 'reviewerID', 'vec', 'overall', 'user_topics_cluster', 'cluster_category_consume_proportion']]
low_consuming = high_consuming[high_consuming['cluster_category_consume_proportion']<=10]
high_consuming = high_consuming[high_consuming['cluster_category_consume_proportion']>=20]

high_consuming_mean_overral = high_consuming['overall'].mean()
low_consuming_mean_overral = low_consuming['overall'].mean()

print(f"Avaliação média de categoria com consumo alto ou regular do cluster: {high_consuming_mean_overral}")
print(f"Avaliação média de categoria com consumo baixo do cluster: {low_consuming_mean_overral}")
```
